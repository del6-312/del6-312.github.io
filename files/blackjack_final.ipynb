```python
def deal_blackjack_hands(deck: list) -> list:
  """
  Deals two hands for blackjack: one for dealer and another for the player.

  Args:
                   deck (list):           A list representing a deck of cards.

  Returns:
      final_hands (list):                 List containing the hands that were dealt. Each hand is a list of cards. The first hand is the dealer's and the other is the player's

  Examples:
  >>> deal_blackjack_hands(create_deck())
      [['As', '2s'], ['3s', '4s']]
  """
  # Will count how many cards are dealt so far
  count = 1
  # Holds an individual hand at once
  hand_holder = []
  # Holds all of the individual hands
  final_hands = []

  # Will deal 4 total cards (two hands of two)
  for i in range(4):
    # When there are not two cards, add the card to the hand holder
    if count % 2 != 0:
      hand_holder += [deck.pop(0)]
      count += 1
    # When there are exactly two cards, add the card to the hand holder, and add the hand to the final hand holder
    else:
      # Adding card to the hand holder
      hand_holder += [deck.pop(0)]
      # Adding the hand to the final return
      final_hands += [hand_holder]
      # Empties hand holder for the next hand
      hand_holder = []
      # Adds count to indicate that a card has been added
      count += 1

  # Returning the final hands
  return final_hands
```

```python
def input_money(user_balance: int) -> int:
    """
    Function will asks how much money the user will bet.

    Args:
        user_balance (int):      An integer representing the number of dollars the user currently has.

    Returns:
          bet_amount (int):      An integer representing the number of dollars the user would like to spend
    """
    # Prints the current balance of the user
    print(f"Current Balance: ${user_balance}")

    # Asks the user how much they will bet
    bet_amount = input("Place your bet: $")

    # If the bet is not a number you need to keep on asking for a value until a proper number is given.
    while not bet_amount.isnumeric():
        # Asks user to give another input
        bet_amount = input("Invalid input. Please type in a number: $")

    # Once user inputted a number, the bet amount is turned into an integer.
    bet_amount = int(bet_amount)

    # Now, if your bet is larger than than the amount of balance you have, then you need to keep on putting in a input that is valid.
    while bet_amount > user_balance:
        # Asks for another valid input
        bet_amount = int(input("Insufficent balance, please place a lower bet: $"))


    # If all the qualifications have been met, then you return the bet amount
    return bet_amount
```

```python
def evaluation(player_hand: list, dealer_hand: list, deck: list) -> int:
    """
    Will present the initial hands of both the dealer and the user. It will also evaluate the user's and dealer's hands after hitting and standing. It will also evaluate who won.

    Args:
        player_hand (int):        Represents the player's initial hand dealt
        dealer_hand (int):        Represents the dealer's initial hand dealt
              deck (list):        Represents the deck (after the player's and dealer's hands have been dealt)

    Returns:
                    (int):        Returns 1 if the user won. Returns 0 is the user lost. Returns 2 if user tied with dealer
    """
    # Holds the value of the dealer's hand (Which is only one card at first)
    dealer_value = 0

    # If the dealer's first card is a Jack, Queen, King, or Ten, then you need to add ten (The index looks weird because I have to find the number within the cards)
    if dealer_hand[0][11] in "JQKT":
        dealer_value = 10

    # If the dealer has an ace, then the dealer has 1/11 (only one card being shown at first)
    elif dealer_hand[0][11] == "A":
        dealer_value = "1/11"

    # In the case the dealer has a card with an integer, the value is just the integer
    else:
        dealer_value = int(dealer_hand[0][11])



    # Holds the value of the player's hand
    player_value = 0

    # Looking through each of the user's cards
    for element in player_hand:
        # If the card is a Jack, Queen, King, or Ten, then you add 10
        if element[11] in "JQKT":
            player_value += 10

        # If the card is an Ace, add one for now. The duality of an ace will be evaluated later once all cards values are added
        elif element[11] == "A":
            player_value += 1

        # If the card is a number card, add the number
        else:
            player_value += int(element[11])


    # Evaluation process for the duality of Aces
    # If the player's hand contains an A and the total value is less than 11, then an ace can act as 1 or 11.
    if (player_hand[0][11] == "A" or player_hand[1][11] == "A") and player_value < 11:
        # Need to show the two possible values with the ace being a 1 or 11
        player_value = f"{player_value}/{player_value + 10}"

    # If the player's hand contains an A and the total value is 11, then you have 21, You've already won
    elif (player_hand[0][11] == "A" or player_hand[1][11] == "A") and player_value == 11:
        player_value = 21
    # In every other case, the value of the ace will stay as 1, as it was initally set



    # Prints the intial results from the deal
    print(f"Dealer's hand:\n{dealer_hand[0]}, --- ({dealer_value})")
    print()
    print(f"Player's hand:\n{player_hand[0]},\n{player_hand[1]} ({player_value})")


    # In the case the user possibly had a blackjack at the start, you need to show that the user won
    if player_value == 21:
        # Print that user won
        print("You Win!")
        # If user types yes, then you will return 1 (which represents a win for this function)
        if input("If you would like to play again, type 'Yes'\n").upper() == "YES":
            return 1


    # Player now can choose to hit, stand
    player_choice = input("Choose a move (S for Stand or H for Hit): ").upper()
    # If player didn't select S or H, then the choice is invalid, and you need ask for a choice until the choice is valid
    while player_choice not in ["S", "H"]:
        player_choice = input("Invalid Choice. Please choose a move (S for Stand or H for Hit): ").upper()

    # If the player hits, use the function that simulates hitting
    if player_choice == "H":
        # Stores the results from the function
        results = hitting(player_hand, dealer_hand, deck, player_value, dealer_value, player_choice)

        # If the first element of the results turns out to be an integer, then the player busted and lost
        if type(results[0]) == int:
            # Return 0 to indicate that player lost
            return 0

        # In the other case, you need to update the variables to the variables that were returned
        else:
            # The player's hand after hitting
            player_hand = results[0]

            # The updated deck after hitting
            deck = results[1]

            # The player's updated value after hitting
            player_value = results[2]

            # The player's updated choice (which only happens when player stands)
            player_choice = results[3]

    # In the end, all of the values will be assesed and it will return if the player lost or won
    return standing(player_hand, dealer_hand, deck, player_value, dealer_value)
```

```python
def hitting(player_hand: list, dealer_hand: list, deck: list, player_value: int, dealer_value: int, player_choice: str) -> tuple:
    """
    Simulates when the player chooses to hit in blackjack. If the user exceeds 21, then the user busts

    Args:
         player_hand (list):       A single list parameter that represents the user's initial hand
         dealer_hand (list):       A single list parameter that represents the dealer's intial hand (which is initially only the first card)
         player_value (int):       A single integer parameter that represents the user's value based on their hand
         dealer_value (int):       A single integer parameter that represents the dealer's value based on the dealer's hand (which is initally only the first card)
        player_chocie (str):       A single string parameter that represents the player's choice (standing or hitting)

    Returns:
                    (tuple):       Returns a tuple with an integer is player busted. Returns a tuple with the player hand, deck, player value, and player choice otherwise
    """
    # This variable will eventually mark if an ace was present in the user's hand. It will be used in specific cases
    ace_marker = int()

    # Variable that holds the player's hand as a string (used more for display)
    player_line = f"Player's hand:\n{player_hand[0]},\n{player_hand[1]}"

    # Loops through a process until the user decides to stand
    while player_choice == "H":

        # Clear's the output (aesthetic purposes)
        clear_output()

        # When the user hit's, a card will be selected from the shuffled deck
        hitting_card = deck.pop(0)

        # This card will be added to the player's hand
        player_hand.append(hitting_card)

        # The player's value now has to be updated too
        # If the player's value is a string (meaning there's an ace that can act as 1/11), then you will go into a conditional statement
        if type(player_value) == str:
            # First marks that there is an ace
            ace_marker = 1
            # Sets the player's value as the lower value for now
            player_value = int(player_value[0])


        # Now you need to evaluate the player's value after you add the hitting card
        # If the card is a Jack, Queen, King, or Ten, then you add 10
        if hitting_card[11] in "JQKT":
            player_value += 10

        # If the card is an Ace, then you need to evaluate what your value will be
        elif hitting_card[11] == "A":
            # If the player's current value is 10, then adding an ace will automatically get you 21
            if player_value == 10:
                player_value = 21
            # If the player's value is less than ten, then ace can act as 1 or 11
            elif player_value < 10:
                player_value = f"{player_value + 1}/{player_value + 11}"
            # In every other case, ace just acts as 1
            else:
                player_value += 1

        # If an ace is present already in the hand (before hitting), then you want to make sure the value is correctly represented
        elif ace_marker == 1:
            # if the hitting card plus the player's lower value is less than or equal to 11, then that means that ace can act as 1 or 11
            if int(hitting_card[11]) + player_value <= 11:
                player_value = f"{player_value + int(hitting_card[11])}/{player_value + int(hitting_card[11]) + 10}"
            # In every other case, ace continues to stay as the lower value
            else:
                player_value += int(hitting_card[11])

        # If the card is a number card, just add the number
        else:
            player_value += int(hitting_card[11])


        # Printing dealer's hand again (because you cleared output earlier)
        print(f"Dealer's hand:\n{dealer_hand[0]}, --- ({dealer_value})")
        # Making blank line
        print()
        # Adding the newly hit card to the variable that holds the viewer display
        player_line += f",\n{hitting_card}"
        # Will print the line along with the value (value is added indepedently so that you dont have values repeated within the variable player_line)
        print(player_line + f" ({player_value})")

        # If the player value is a string (because of an ace), then you want to change it to an integer so that you can evaluate if the player busted or not.
        # If an ace can act as a 1 or 11, then that means that player can't bust anyways, so you can just set it as a random integer for now before you evaluate if player busted.
        if type(player_value) == str:
            player_value = int(player_value[0])

        # See's if player busted or not
        if player_value > 21:
            print("Player Busts! You Lose.")
            if input("If you would like to play again, type 'Yes'\n").upper() == "YES":
                # Returns that player lost (I made it a tuple so that my typehints would always return one data type)
                return (0,)

        # If player didn't bust, then you ask if player will choose to stand or hit
        else:
            player_choice = input("Choose a move (S for Stand, H for Hit): ").upper()
            # If the play doens't give a valid response, then you need to keep on asking until you get a valid response
            while player_choice.upper() not in ["S", "H"]:
                player_choice = input("Invalid Choice. Please choose a move (S for Stand, H for Hit, or DD for Double Down)")

    # If the player selected to stand, then you leave the while loop and you will return the updated deck, player hand, player value, and player choice (which is stand)
    return player_hand, deck, player_value, player_choice
```

```python
def standing(player_hand: list, dealer_hand: list, deck: list, player_value: int, dealer_value: int) -> int:
    """
    Will simulate the process after the player stands (dealer hitting and winner being decided).

    Args:
        player_hand (list):       A single list parameter representing the player's hand after the player decided to stand
        dealer_hand (list):       A single list parameter representing the dealer's hand before the dealer hits
               deck (list):       A single list parameter representing the deck after dealing hands and the player hitting
        player_value (int):       A single integer parameter representing the player's value after hitting (if the player hit)
        dealer_value (int):       A single integer parameter representing the dealer's value.

    Returns:
                     (int):       Will return 1 if player won. Will return 0 if player lost. Will return 2 if player and dealer tied
    """
    # If the dealer's first card is an ace, set it as the higher value for now
    if dealer_hand[0][11] == "A":
        dealer_value = 11

    # If the dealer's second card is a Jack, Queen, King, or Ten, then you add 10 to the dealer value
    if dealer_hand[1][11] in "JQKT":
        dealer_value += 10

    # If the dealer's second card is an ace, then you have multiple cases
    elif dealer_hand[1][11] == "A":
        # If the dealer's value is 10, then dealer has a blackjack
        if dealer_value == 10:
            dealer_value == 21
        # If the dealer value is greater than 10, then the ace acts as an 11
        elif dealer_value < 10:
            dealer_value += 11
        # In every other case, ace acts as an 11
        else:
            dealer_value += 1
    # If the card is a number card, add the number
    else:
        dealer_value += int(dealer_hand[1][11])


    # Creating a dealer line, which will be displayed later in the game (it will update based on if the dealer has to hit or not)
    dealer_line = f"Dealer's hand:\n{dealer_hand[0]},\n{dealer_hand[1]}"

    # The dealer will keep on hitting until their value is at 17 or above
    while dealer_value < 17:

        # Clearing output for aesthetic purposes
        clear_output()

        # Getting the card that the dealer will hit
        hitting_card = deck.pop(0)

        # Adding that card to the dealer's hand
        dealer_hand.append(hitting_card)

        # If the card is a Jack, Queen, King, or Ten, then you add 10 to the dealer value
        if hitting_card[11] in "JQKT":
            dealer_value += 10

        # If the card is an Ace, there are multiple possibilities
        elif hitting_card[11] == "A":
            # If the dealer value is equal to 10, then the ace will act as 11 (dealer will have 21)
            if dealer_value == 10:
                dealer_value == 21
            # If the dealer's value is less than 10, then you will add 11
            elif dealer_value < 10:
                dealer_value += 11
            # If dealer's value is greater 10, then you just add 1 for the ace
            else:
                dealer_value += 1

        # If the card is a number card, add the number
        else:
            dealer_value += int(hitting_card[11])

        # Add the new card to dealer line (visual representation)
        dealer_line += f",\n{hitting_card}"

    # Setting up the user's values before dealer and user values are compared
    # If user value is a string (meaning ace is acting as 1 or 11), use the higher value
    if type(player_value) == str:
        # Last to characters of list are the higher values
        player_value = int(player_value[-2:])

    # Clearing any output if there is possibly any (just in case)
    clear_output()
    # Printing the dealer line with the dealer's value (visual representation)
    print(dealer_line + f" ({dealer_value})")
    # Printing empty line
    print()

    # Setting up player line so it can be printed. Intially only contains first two cards because those are the ones that will definately always be there (more than 2 aren't promised)
    player_line = f"Player's hand:\n{player_hand[0]},\n{player_hand[1]}"
    # Will add any extra cards that were hit to the player line
    for i in range(len(player_hand) - 2):
        player_line += f",\n{player_hand[i + 2]}"
    # Adds the player value at the end
    player_line += f" ({player_value})"
    # Prints the player line
    print(player_line)


    # If the dealer busts, then player wins
    if dealer_value > 21:
        print("Dealer Busts! Player Wins!")
        if input("If you would like to play again, type 'Yes'\n").upper() == "YES":
            # Return 1 to repesent that player won
            return 1

    # Player value and dealer value will be compared. The result of the comparison will be returned
    return standoff(player_value, dealer_value)
```

```python
def standoff(player_value: int, dealer_value: int) -> int:
    """
    Will compare user and dealer values to see if player lost or won

    Args:
        player_value (int):         A single integer parameter representing the value of the player's hand
        dealer_value (int):         A single integer parameter representing the value of the dealer's hand

    Returns:
                     (int):         Will return 1 if player won. Will return 0 if player lost. Will return 2 is player and dealer tied

    Examples:
     >>>standoff(18, 17)
        1

     >>>standoff(14, 20)
        0

     >>>standoff(20, 20)
        2
    """
    # If the player has a higher value than the dealer, then the player won
    if player_value > dealer_value:
        print("Player Wins!")
        if input("If you would like to play again, type 'Yes'\n").upper() == "YES":
            # 1 represents that player won
            return 1

    # if the dealer has a higher value, then the dealer won
    elif player_value < dealer_value:
        print("Player Loses.")
        if input("If you would like to play again, type 'Yes'\n").upper() == "YES":
            # return 0 to indicate that player lost
            return 0

    # In every other case, the dealer and user tied. You need to return that dealer and player tied
    else:
        print("Push (Dealer and Player tied)")
        if input("If you would like to play again, type 'Yes'\n").upper() == "YES":
            # Returns 2 to indicate that there was a tie
            return 2
```
